```python
from urllib.parse import urlparse, urlunparse

def sanitize_url(url: str) -> str:
    url = url.strip()                    # remove whitespace
    
    if not url.startswith(("http://", "https://")):
        url = "https://" + url           # add scheme if missing
    
    parsed = urlparse(url)
    
    clean = parsed._replace(
        scheme=parsed.scheme.lower(),    # https://EXAMPLE.com -> https://
        netloc=parsed.netloc.lower(),    # EXAMPLE.COM -> example.com
        fragment=""                      # remove #section parts
    )
    
    return urlunparse(clean)
```
